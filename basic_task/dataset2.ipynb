```python
import numpy as np
import pandas as pd
```

```python
# These are the headers which you need to link to your dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv('car_dataset.data', names = headers)
```

```python

df.head(15)
```

```python



df.replace("?",np.nan,inplace=True)
```

```python
df.isnull()
```

```python
missing_values=df.isnull()
missing_values.head(5)
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
df["bore"].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_horse_power = df["horsepower"].astype("float").mean(axis=0)
df["horsepower"].replace(np.nan,avg_horse_power,inplace = True)
```

```python
avr_rpm = df["peak-rpm"].astype("float").mean(axis=0)
```

<h4>Replace NaN by mean value:</h4>


```python
df["peak-rpm"].replace(np.nan,avr_rpm,inplace =True)
```

```python
freq = df["num-of-doors"].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan,freq,inplace = True)
```

```python


df.dropna(subset=['price'],axis=0,inplace=True)

df.reset_index(drop=True,inplace=True)
```

```python
numrical_data = df.select_dtypes(include="number")
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
correlated_data = numrical_data.corr()
# print()

        
mul = correlated_data.values[0]*correlated_data

plt.figure(figsize=(16,14))
sb.heatmap(mul,annot=True)
plt.show()
sum = 0.0
for i in range(0,10):

    sum += mul.values[1][i]
```
